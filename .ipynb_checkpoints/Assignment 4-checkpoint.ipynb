```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.cluster import DBSCAN
from sklearn.cluster import SpectralClustering
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from sklearn.datasets import make_moons
from matplotlib.colors import ListedColormap
import sklearn.metrics as metrics
```

```python
def k_means(k,X):
    thresh = 0.1
    # generate ramdom means (selecting obsevations)
    u = np.zeros((k,len(X[0])))
    #for i in range(k):
        #for j in range(len(X[0])):
            #u[i,j] = np.random.uniform(min(X[:,j]), max(X[:,j]))
    rand = [i for i in range(len(X))]
    sample = np.random.choice(rand, size=k, replace=False)
    for i in range(k):
        u[i,:] = X[sample[i],:]
    C_temp = 0
    while True:
        #calculate dist
        dist = np.zeros((len(X),k)) #l2_norm
        for i in range(k):
            for j in range(len(X)):
                d = X[j,:] - u[i,:]
                dist[j,i] = d.dot(d)
        #assign points
        r = np.zeros((len(X),k)) #r_ik
        for i in range(len(X)):
            r[i,0] = 1
            temp = 0
            for j in range(k-1):
                if dist[i,j+1] < dist[i, temp]:
                    r[i,temp] = 0
                    r[i,j+1] = 1
                    temp = j+1
        #update u
        for i in range(k):
            for j in range(len(X[0])):
                u[i,j] = np.mean(X[:,j][r[:,i]==1])
        #calculate C (Sum of squared err)
        C = np.sum(np.diag(r.T.dot(dist)))
        #print(C)
        if C_temp - C < thresh and C_temp > 0:
            break
        else:
            C_temp = C
    #return array of label
    lab = 0 * r[:,0]
    for i in range(k):
        lab = lab + i * r[:,i]
    return lab, C
```

**Explain in 1-2 sentences each (and at a very high level) how the following clustering techniques work and what distinguishes them from other clustering methods: (1) k-means, (2) agglomerative clustering, (3) Gaussian mixture models, (4) DBSCAN, and (5) spectral clustering**

(1) k-means: works by finding the centroid of each cluster that minimize the sum of square error in an iterative manner where data points are assigned to the nearest cluster. Number of cluster need to be determined upfront.

(2) agglomerative clustering: each data point starts as a unique cluster and pairs of clusters are merged based on selected measure of dissimilarity and linkage as one moves up the dendrogram. The dendrogram can be cut at certain point to produce desired number of clusters.

(3) Gaussian mixture models: works by finding the probability distribution of each cluster that maximize the likelyhood in an iterative manner. Relaxs the equal and diagonal covariance assumption of k-means.

(4) DBSCAN: works by marking points as core, border, or noise points according to number of nearby neighbours and grouping neighbouring core and border points as a cluster. It produces outliners that fall into none of the classes.

(5) spectral clustering: uses eigenvalues of the similarity matrix of the data to perform dimensionality reduction before clustering in fewer dimensions with other methods such as K-means. focuses on connectedness instead of compactness of the points.

(a)

**Create a neural network class that follows the scikit-learn classifier convention by implementing fit, predict, and predict_proba methods. Your fit method should run backpropagation on your training data using stochastic gradient descent. Assume the activation function is a sigmoid.**

```python
def sig(x): #sigmoid
    return 1 / (1 + np.exp(-x))
def sig_d(y): #derivative of sigmoid
    return y * (1 - y)
```

```python
# 1. Batch gradient descent
class Neural_Network():
    def __init__(self, node=5, thresh=0.00000001, max_i=50000):
        self.node = node
        self.layer = 2
        self.lr = 0.1 #learning rate
        self.thresh = thresh
        self.max_i = max_i

    def fit(self, X, Y):
        np.random.seed(1)
        weight_1 = 2 * np.random.random((2, self.node)) - 1 
        weight_2 = 2 * np.random.random((self.node, self.node)) - 1
        weight_3 = 2 * np.random.random((self.node, 1)) - 1
        Err = 0
        for i in range(self.max_i):
            Z_1 = sig(np.dot(X, weight_1)) #n*5
            Z_2 = sig(np.dot(Z_1, weight_2)) #n*5
            y_hat = sig(np.dot(Z_2, weight_3)) #n*1
            Err_i = 0.5 * np.sum((Y - y_hat)**2) / len(X)
            dErr_3 = (Y - y_hat) * sig_d(y_hat)
            dErr_2 = np.dot(dErr_3, weight_3.T) * sig_d(Z_2)
            dErr_1 = np.dot(dErr_2, weight_2.T) * sig_d(Z_1)
            weight_1 += self.lr * np.dot(X.T, dErr_1) 
            weight_2 += self.lr * np.dot(Z_1.T, dErr_2)
            weight_3 += self.lr * np.dot(Z_2.T, dErr_3)
            if Err - Err_i < self.thresh and Err > Err_i:
                break
            Err = Err_i
        self.weight_1 = weight_1
        self.weight_2 = weight_2
        self.weight_3 = weight_3
    
    def predict(self, X_test):
        thresh = 0.5
        a = sig(np.dot(X_test, self.weight_1))
        a = sig(np.dot(a, self.weight_2))
        a = sig(np.dot(a, self.weight_3))
        return (a > thresh)*1
    
    def predict_proba(self, X_test):
        a = sig(np.dot(X_test, self.weight_1))
        a = sig(np.dot(a, self.weight_2))
        a = sig(np.dot(a, self.weight_3))
        return a
```

```python
# 2. Stochastic gradient descent
class Neural_Network_Stochastic():
    def __init__(self, node=5, thresh=0.000001, max_i=5000):
        self.node = node
        self.layer = 2
        self.lr = 0.1 #learning rate
        self.thresh = thresh
        self.max_i = max_i

    def fit(self, X, Y):
        p = len(X[0])
        np.random.seed(1)
        weight_1 = 2 * np.random.random((p, self.node)) - 1 
        weight_2 = 2 * np.random.random((self.node, self.node)) - 1
        weight_3 = 2 * np.random.random((self.node, 1)) - 1
        Err = 0
        data = np.c_[X, Y]
        for i in range(self.max_i):
            np.random.shuffle(data)
            Err_n = 0
            for j in range(len(data)):
                X = data[j, 0:p].reshape((1,p))
                Y = data[j, p]
                Z_1 = sig(np.dot(X, weight_1)) #n*5
                Z_2 = sig(np.dot(Z_1, weight_2)) #n*5
                y_hat = sig(np.dot(Z_2, weight_3)) #n*1
                dErr_3 = (Y - y_hat) * sig_d(y_hat)
                dErr_2 = np.dot(dErr_3, weight_3.T) * sig_d(Z_2)
                dErr_1 = np.dot(dErr_2, weight_2.T) * sig_d(Z_1)
                weight_1 += self.lr * np.dot(X.T, dErr_1) 
                weight_2 += self.lr * np.dot(Z_1.T, dErr_2)
                weight_3 += self.lr * np.dot(Z_2.T, dErr_3)
                Err_n += 0.5 * (Y - y_hat)**2
            if Err - Err_n < self.thresh and Err > Err_n:
                break
            Err = Err_n
        self.weight_1 = weight_1
        self.weight_2 = weight_2
        self.weight_3 = weight_3
    
    def predict(self, X_test):
        thresh = 0.5
        a = sig(np.dot(X_test, self.weight_1))
        a = sig(np.dot(a, self.weight_2))
        a = sig(np.dot(a, self.weight_3))
        return (a > thresh)*1
    
    def predict_proba(self, X_test):
        a = sig(np.dot(X_test, self.weight_1))
        a = sig(np.dot(a, self.weight_2))
        a = sig(np.dot(a, self.weight_3))
        return a
```

```python
train_X, train_Y = make_moons(500, noise=0.20)
train_Y_r = train_Y.reshape((500,1))
test_X, test_Y = make_moons(100, noise=0.20)
```

```python
#plot
cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000',  '#0000FF'])
xx,yy = np.meshgrid(np.arange(-2,3,0.01),np.arange(-1.5,2,0.01))
X_b = np.c_[xx.ravel(),yy.ravel()]
b_res = nn.predict(X_b).reshape(xx.shape)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.pcolormesh(xx, yy, b_res, cmap=cmap_light)
plt.scatter(train_X[:,0], train_X[:,1], c=train_Y, cmap = cmap_bold, marker = '.')
plt.title("training data")
plt.xlabel("X1")
plt.ylabel("X2")
plt.subplot(1,2,2)
plt.pcolormesh(xx, yy, b_res, cmap=cmap_light)
plt.scatter(test_X[:,0], test_X[:,1], c=test_Y, cmap = cmap_bold, marker = '.')
plt.title("test data")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

*Figure 9. dataset and decision boudary as a result of neural network classifier with 2 hidden layer and 5 nodes for each layer. (left) training data, (right) test data*

```python
def make_roc(y, score):
    fpr, tpr, _ = metrics.roc_curve(y, score, pos_label=1)
    auc = metrics.roc_auc_score(y, score)
    legend_string = 'AUC = {:0.3f}'.format(auc)
    plt.plot([0,1],[0,1],'--', color='gray', label='Chance')
    plt.plot(fpr, tpr, label=legend_string)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.grid('on')
    plt.axis('square')
    plt.legend()
    plt.tight_layout()

#make ROC
pred = nn.predict_proba(test_X)
make_roc(test_Y, pred)
plt.show()
```

*Figure 10. ROC curve, blue line is the performance of the neural network classifier on test data, dashed grey line is the performance of classifying by chance*
